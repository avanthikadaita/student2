```python
%%html

<form>
    <p><label>
        Name:
        <input type="text" name="name" id="name" required />
    </label></p>
    <p><label>
        Instrument:
        <input type="text" name="instrument" id="instrument" required />
    </label></p>
    <p><label>
        Time Practiced (minutes):
        <input type="number" name="practiceTime" id="practiceTime" required />
    </label></p>
    <p>
        <button type="button" onclick="practice_log()">Save</button>
    </p>
</form>
<div id="pointsDisplay"></div>
<script>
    let totalPracticeTime = 0; // Variable to store total practice time
    let rewardPoints = 0; // Variable to store reward points

    function practice_log() {
        const name = document.getElementById("name").value;
        const instrument = document.getElementById("instrument").value;
        const practiceTime = parseInt(document.getElementById("practiceTime").value);
        let earnedPoints = 0;

        // Update the total practice time
        totalPracticeTime += practiceTime;

        // Check for reward milestones
        if (totalPracticeTime >= 30 && rewardPoints < 10) {
            rewardPoints = 10;
            earnedPoints = 10;
        }
        if (totalPracticeTime >= 60 && rewardPoints < 20) {
            rewardPoints = 20;
            earnedPoints = 20;
        }
        if (totalPracticeTime >= 120 && rewardPoints < 30) {
            rewardPoints = 30;
            earnedPoints = 30;
        }

        // Display earned points
        displayPointsEarned(earnedPoints);

        // Create the formData object to send to the server
        const formData = { name, instrument, practiceTime, rewardPoints };

        // Send data to the server (Replace the URL with your server endpoint)
        fetch('https://ikan2025.github.io/frontend/', {
            method: 'POST',
            headers: {
                'Content-Type': 'application/json',
            },
            body: JSON.stringify(formData),
        })
        .then(response => response.json())
        .then(data => {
            console.log(data.message);
        })
        .catch(error => {
            console.error('Error:', error);
        });

        // Reset only the practiceTime input
        document.getElementById("practiceTime").value = "";
    }

    function displayPointsEarned(points) {
        const pointsDisplay = document.getElementById("pointsDisplay");
        pointsDisplay.innerText = `You've earned ${points} points!`;
    }
</script>
```
